# Time Series Analysis of Liquid Water Equivalent (LWE)

This notebook demonstrates how to analyze the Liquid Water Equivalent (LWE) timeseries data. The data is read, processed, and visualized to explore trends in groundwater levels over time.

## 1. Importing Required Libraries

First, we need to import the necessary libraries for data manipulation and visualization.

````python
import pandas as pd
import matplotlib.pyplot as plt
````

## 2. Loading the Dataset

Next, we load the dataset from a CSV file. The `date` column is converted into a `datetime` format for easier manipulation, and we set the `date` column as the index of the DataFrame for time-based plotting.


````python
path=''

df=pd.read_csv(path)

df['date']=pd.to_datetime(df['date'])

df=df.set_index('date')
````

## 3. Plotting the Liquid Water Equivalent (LWE) Time Series

Now, let's visualize the LWE data by plotting it as a time series. We will create a line plot to observe how the LWE values change over time.


````python


plt.figure(figsize=(8,4))
plt.rcParams['font.size']=12
plt.rcParams['font.family']='Arial'
ax=plt.subplot(111)
plt.plot(df)
plt.xlabel('Date')
plt.ylabel('Liquid Water Equivalent Thickness (cm)')
plt.title('LWE Timeseries for Khorda')
plt.tight_layout()

ax.spines[['top','right']].set_visible(False)
````

## 4. Adding a Trendline to the Time Series

In this section, we will add a linear regression trendline to the Liquid Water Equivalent (LWE) time series data. The goal is to visualize the underlying trend over time. We will use **`scikit-learn`**'s `LinearRegression` model to fit a line to the data and display the trend.

### Steps Involved:
1. **Prepare Data for Linear Regression**: We convert the datetime index of the time series to numerical values (representing the number of days since the start).
2. **Fit the Linear Regression Model**: Using the numerical data (`x`) and LWE values (`y`), we fit a linear regression model.
3. **Generate the Trendline**: Once the model is fitted, we generate the predicted values (the trendline).
4. **Plot the Trendline**: Finally, the trendline is overlaid on the original time series plot, and we display the trendline equation on the graph.

````python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Assuming 'df' is your DataFrame with a datetime index and the time series data

# Create a figure and axis
plt.figure(figsize=(8, 4))
plt.rcParams['font.size'] = 12
plt.rcParams['font.family'] = 'Arial'

ax = plt.subplot(111)

# Plot the time series data
plt.plot(df, label='LWE Time Series')

# Prepare the data for linear regression
# Convert the datetime index into numerical values (e.g., days since the start)
x = np.arange(len(df)).reshape(-1, 1)  # x should be a 2D array (required by sklearn)
y = df.values  # y is the values of the time series

# Create the LinearRegression model
model = LinearRegression()

# Fit the model
model.fit(x, y)

# Get the slope (coefficient) and intercept of the trendline
slope = model.coef_[0]
intercept = model.intercept_

# Generate the trendline using the model
trendline = model.predict(x)

# Plot the trendline
plt.plot(df.index, trendline, color='red', linestyle='--', label='Trendline')

# Labeling
plt.xlabel('Date')
plt.ylabel('Liquid Water Equivalent Thickness (cm)')
plt.title('LWE Timeseries for Khorda')

# Add the trendline equation and value as text on the plot
equation = f'Trendline: y = {slope}x + {intercept}'
plt.text(0.5, 0.95, equation, transform=ax.transAxes, color='red', fontsize=12, 
         verticalalignment='top', horizontalalignment='left')

# Make the plot look nicer
plt.tight_layout()

# Hide the top and right spines
ax.spines[['top', 'right']].set_visible(False)


# Display the plot
plt.show()
````

## 6. Enhanced Boxplot Visualization of Liquid Water Equivalent (LWE) Variations

In this section, we will create a more visually appealing boxplot to compare the Liquid Water Equivalent (LWE) thickness variations for three datasets: CSR, GFZ, and JPL. The boxplot will help us understand the distribution and variability of LWE across these datasets, while enhanced aesthetics will make the plot more readable and informative.

### Key Features of the Enhanced Boxplot:
1. **Custom Box Colors**: The boxes are filled with light blue to make the plot more visually appealing.
2. **Black Median Line**: The median line is shown in black to enhance contrast.
3. **Customized Whiskers and Caps**: The whiskers and caps are also black, with thicker lines for better visibility.
4. **Gridlines**: Horizontal gridlines are added to make it easier to compare the values across the boxes.
5. **Axis Labels and Title**: The axis labels and title are customized for clarity.

### Steps Involved:
1. **Data Preparation**: We load the data and ensure the 'date' column is converted to the correct datetime format. The 'date' column is then set as the index for easier handling.
2. **Boxplot Creation**: The boxplot is created using `matplotlib` with customized aesthetic properties.
3. **Axis Customization**: We adjust the labels, title, and add gridlines to improve the readability of the plot.
4. **Plot Display**: Finally, we remove unnecessary spines and apply a tight layout for a polished output.


````python

import pandas as pd
import matplotlib.pyplot as plt

# Read data
df = pd.read_csv('')

# Convert 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'], format="%d-%m-%Y")
df = df.set_index('date')

# Create figure and axis
plt.figure(figsize=(6,4))
plt.rcParams['font.size'] = 11  # Increase font size for better readability
plt.rcParams['font.family'] = 'Arial'

# Define axis
ax = plt.subplot(111)

# Create the boxplot with enhanced aesthetics
box = plt.boxplot([df['lwe_thickness_csr'], df['lwe_thickness_gfz'], df['lwe_thickness_jpl']],
                  patch_artist=True,  # Fill the box with color
          
                  widths=0.6,         # Adjust box width
                  medianprops={'color': 'black', 'linewidth': 2},  # Black median line
                  boxprops=dict(facecolor='lightblue', color='black'),  # Box color
                  whiskerprops=dict(color='black', linewidth=1.5),  # Whisker color and width
                  capprops=dict(color='black', linewidth=1.5))  # Cap color and width

# Customize xticks and labels
plt.xticks([1, 2, 3], ['CSR', 'GFZ', 'JPL'], fontsize=12)
plt.ylabel('Liquid Water Equivalent \n Thickness Variation (cm)', fontsize=14)

# Title with adjusted font size
plt.title('LWE Timeseries Variation for Khorda', fontsize=16)

# Add gridlines for better readability
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Remove top and right spines for cleaner look
ax.spines[['top', 'right']].set_visible(False)

# Add a tight layout
plt.tight_layout()

# Show the plot
plt.show()

````
